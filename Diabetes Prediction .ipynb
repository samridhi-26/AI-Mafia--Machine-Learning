```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfxt = pd.read_csv('Diabetes_XTrain.csv')
dfy = pd.read_csv('Diabetes_YTrain.csv')
dfx = pd.read_csv('Diabetes_Xtest.csv')
```

```python
XTR = dfxt.values
X = dfx.values
Y = dfy.values

print(X.shape)
print(Y.shape)
```

```python
def distance(p1,p2):
    return np.sum((p2 - p1)**2)**.5

def knn(X,Y,test,k=5):
    m = X.shape[0]
    
    vals = []
    for i  in range(m):
        dist = distance(test,X[i])
        vals.append((dist,Y[i]))
    
    vals = np.array(sorted(vals))[:,1]
    vals = vals[:k]
    t =  np.unique(vals,return_counts=True)
    idx = np.argmax(t[1])
    pred = int(t[0][idx])
        
    return pred
    
```

```python
diabetes ={"RESULT": []}
```

```python
for i in range(X.shape[0]):
    test_data = X[i]
    ans = knn(XTR,Y,test_data,k=17)
    diabetes["RESULT"].append(ans)
```

```python
print(diabetes)
```

```python
prediction = pd.DataFrame(diabetes)
prediction.to_csv('prediction.csv',index = False)
```
